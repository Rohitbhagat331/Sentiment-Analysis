```python
import os
from nltk import word_tokenize
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer ,PorterStemmer
from nltk.corpus import stopwords
import re
from string import digits
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(comment):
    try:
        comment = str(comment)
        comment = comment.lower()
        clear = re.compile('[-+]?(0[xX])?[\dA-Fa-f➀ ② ② ➀]+[-+]?(0[xX][\dA-Fa-f]+|0[0-7]*|\d+)[-+]?\d+(\.\d*)?:=;[A-Za-z]+\n[ \t]+.')
        clear_text = re.sub(clear,'',comment)
        remove_url = re.sub(r'http\S+','',clear_text)
        remove_number = re.sub('[0-9]+','',remove_url)
        tokenizer = RegexpTokenizer(r'\w+|\$[\d\.]+|\S+[a-z]')
        tokens = tokenizer.tokenize(remove_number)
        filtered_words = [lemmatizer.lemmatize(word) for word in tokens]
        stem_words_remove_pre_suf = [ps.stem(w) for w in filtered_words]
        return " ".join(filtered_words)
    except Exception as e:
        print('error in preprocess function',e)
        
        
```

```python
df1['preprocess_text'] = df['translated_text'].apply(lambda x : preprocess(x))
```

```python
from transformers import pipeline
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_name = "xlm-roberta-base"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
from tqdm import  tqdm
```

```python
Emptyline = []

for row in tqdm(df1['preprocess_text']):
    vk = classifier(str(row),truncation = True)
    Emptyline.append(vk)
    df_sentiments = pd.DataFrame(Emptyline)
     
```
